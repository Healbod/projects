3. Как долго проводилась кросс-валидация для градиентного бустинга с 30 деревьями? Какое качество при этом получилось?

```python
X = features.drop(columns=['duration', 'radiant_win', 'tower_status_radiant', 'tower_status_dire', 'barracks_status_radiant', 'barracks_status_dire'])
y = features['radiant_win']

X = X.fillna(0)
```

```python
import time
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True)
clf = GradientBoostingClassifier(n_estimators=30)
start_time = time.time()

scores = cross_val_score(clf, X, y, scoring='roc_auc', cv=kf)
score = scores.mean()

print("Качество на кросс-валидации по 5 блокам: ", score)
print("Затраченное время: ", time.time() - start_time)
```

4. Имеет ли смысл использовать больше 30 деревьев в градиентном бустинге? Что можно сделать, чтобы ускорить его обучение при увеличении количества деревьев?

```python
#Попробуем взять разное количество деревьев

for n in range(10, 110, 10):
    clf = GradientBoostingClassifier(n_estimators=n)
    start_time = time.time()

    scores = cross_val_score(clf, X, y, scoring='roc_auc', cv=kf)
    score = scores.mean()
    
    print("Количество деревьев: ", n)
    print("Качество на кросс-валидации по 5 блокам: ", score)
    print("Затраченное время: ", time.time() - start_time)
    print("-----------------")

```

Ответ: Имеет смысл использовать больше деревьев, время увеличивается линейно.
Чтобы ускорить время, можно увеличить значение learning_rate или сократить количество признаков

## Подход 2: логистическая регрессия

1. Какое качество получилось у логистической регрессии над всеми исходными признаками? Как оно соотносится с качеством градиентного бустинга? Чем вы можете объяснить эту разницу? Быстрее ли работает логистическая регрессия по сравнению с градиентным бустингом?

```python
#Отключим Future warnings
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

X = features.drop(columns=['duration', 'radiant_win', 'tower_status_radiant', 'tower_status_dire', 'barracks_status_radiant', 'barracks_status_dire'])
X = X.fillna(0)

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

results = []

for k in range(1, 12, 1):
    clf = LogisticRegression(penalty='l2', C=k)
    start_time = time.time()

    scores = cross_val_score(clf, X, y, scoring='roc_auc', cv=kf)
    score = scores.mean()
    
    results.append(score)
    print("Коэффициент регуляризации: ", k)
    print("Качество на кросс-валидации по 5 блокам: ", score)
    print("Затраченное время: ", time.time() - start_time)
    print("-----------------")
    
print('=======================')
print('Лучший результат:', max(results))
```

Ответ: качество у логистической регрессии выше, чем у градиентного бустинга. Это можно объяснить характером данных. Логистическая регрессия обучается намного быстрее, а изменение параметра регуляризации не влияет на время.

2. Как влияет на качество логистической регрессии удаление категориальных признаков (укажите новое значение метрики качества)? Чем можно объяснить это изменение?

```python
#удалим категориальные признаки
X = features.drop(columns=['duration', 'radiant_win', 'tower_status_radiant', 'tower_status_dire', 'barracks_status_radiant', 'barracks_status_dire'])
X = X.fillna(0)
X = X.drop(columns=['r1_hero', 'r2_hero', 'r3_hero', 'r4_hero', 'r5_hero', 'd1_hero', 'd2_hero', 'd3_hero', 'd4_hero', 'd5_hero'])

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

results = []

for k in range(1, 12, 1):
    clf = LogisticRegression(penalty='l2', C=k)
    start_time = time.time()

    scores = cross_val_score(clf, X, y, scoring='roc_auc', cv=kf)
    score = scores.mean()
    
    results.append(score)
    print("Коэффициент регуляризации: ", k)
    print("Качество на кросс-валидации по 5 блокам: ", score)
    print("Затраченное время: ", time.time() - start_time)
    print("-----------------")
    
print('=======================')
print('Лучший результат:', max(results))
```

Ответ: удаление категориальных признаков немного улучшает результат. Это можно объяснить тем, что признак героя в числовом виде вносит искажения.

3. Сколько различных идентификаторов героев существует в данной игре?

Ответ: 113 идентификаторов

4. Какое получилось качество при добавлении "мешка слов" по героям? Улучшилось ли оно по сравнению с предыдущим вариантом? Чем можно это объяснить?

```python
import numpy as np

# N — количество различных героев в выборке
N = 113

X = features.drop(columns=['duration', 'radiant_win', 'tower_status_radiant', 'tower_status_dire', 'barracks_status_radiant', 'barracks_status_dire'])
X = X.fillna(0)

X_pick = np.zeros((X.shape[0], N))

for i, match_id in enumerate(X.index):
    for p in range(5):
        X_pick[i, X.ix[match_id, 'r%d_hero' % (p+1)]-1] = 1
        X_pick[i, X.ix[match_id, 'd%d_hero' % (p+1)]-1] = -1


X_pick_df = pandas.DataFrame(X_pick)
X = X.reset_index()
X = pandas.concat([X, X_pick_df], axis=1)
X
```

```python
X = X.drop(columns=['r1_hero', 'r2_hero', 'r3_hero', 'r4_hero', 'r5_hero', 'd1_hero', 'd2_hero', 'd3_hero', 'd4_hero', 'd5_hero'])

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

results = []
classifiers = []

for k in range(1, 12, 1):
    clf = LogisticRegression(penalty='l2', C=k)
    start_time = time.time()

    scores = cross_val_score(clf, X, y, scoring='roc_auc', cv=kf)
    score = scores.mean()
    
    results.append(score)
    classifiers.append(clf)
    print("Коэффициент регуляризации: ", k)
    print("Качество на кросс-валидации по 5 блокам: ", score)
    print("Затраченное время: ", time.time() - start_time)
    print("-----------------")
    
print('=======================')
print('Лучший результат:', max(results))
```

Ответ: качество значительно улучшается после добавления "мешка слов", так как очевидно, что признак героев вносит существенный вклад

5. Какое минимальное и максимальное значение прогноза на тестовой выборке получилось у лучшего из алгоритмов?

```python
best_model = classifiers[0]
best_model.fit(X, y)
```

```python
test_data = pandas.read_csv('./data/features_test.csv')
features_test = test_data.fillna(0)

features_test_pick = np.zeros((features_test.shape[0], N))

for i, match_id in enumerate(features_test.index):
    for p in range(5):
        features_test_pick[i, features_test.ix[match_id, 'r%d_hero' % (p+1)]-1] = 1
        features_test_pick[i, features_test.ix[match_id, 'd%d_hero' % (p+1)]-1] = -1


features_test_pick_df = pandas.DataFrame(features_test_pick)
features_test = features_test.reset_index()
features_test = pandas.concat([features_test, features_test_pick_df], axis=1)
features_test = features_test.drop(columns=['r1_hero', 'r2_hero', 'r3_hero', 'r4_hero', 'r5_hero', 'd1_hero', 'd2_hero', 'd3_hero', 'd4_hero', 'd5_hero'])

scaler = StandardScaler()
scaler.fit(features_test)
features_test = scaler.transform(features_test)

prediction = best_model.predict_proba(features_test)
```

```python
print("Ответ: ")
print("Максимальное значение: ", prediction.max())
print("Минимальное значение: ", prediction.min())
```
